```python
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("athletes.csv")
```

```python
data.head(3)
```

```python
sns.displot(x=data['weight'],bins=15)
```

```python
sns.displot(x=data['age'],bins=10)
```

```python
sns.jointplot(x=data['age'], y=data['height'], kind='scatter')
sns.jointplot(x=data['age'], y=data['height'], kind='hex')
sns.jointplot(x=data['age'], y=data['height'], kind='reg')
```

```python
sns.jointplot(x=data['weight'], y=data['height'], kind='scatter')
sns.jointplot(x=data['weight'], y=data['height'], kind='hex')
sns.jointplot(x=data['weight'], y=data['height'], kind='reg')
```

```python
sns.scatterplot(x='weight',y='height', data = data)
```

```python
sns.pairplot(data)
```

```python
sns.barplot(x=data['gender'], y=data['height'])
```

```python
sns.barplot(y=data['gender'], x=data['height'])
```

```python
import matplotlib.pyplot as plt
sns.countplot(data['gender'])
plt.savefig('gender.jpg')
```

```python
sns.boxplot(y=data['height'])
```

```python
sns.boxplot(y=data['height'],x=data['gender'])
```

```python
sns.displot(x=data['MPG'], y=data['Horsepower']) #create plot with x axis MPG and y axis Horsepower
plt.title('MPG and Horsepower Comparison for Vehicles') #title 
plt.legend(loc=1)
plt.xlabel('MPG') #x-lable
plt.ylabel('Horsepower')#y-labels
plt.show()
```

```python
sns.displot(x=data['MPG'], y=data['Weight']) # correlation between vehicle weight and MPGs
plt.title('Vehicle Weight and MPG Comparison')#title
plt.xlabel('MPG')#x-label
plt.ylabel('Weight')#ylabel
plt.show()
```

```python
sns.barplot(x=data['MPG'], y=data['Displacement']) #barplot to show MPG and Displacement comparison
plt.title('MPG and Engine Displacement Comparison')
plt.xlabel('MPG')
plt.ylabel('Displacement')
plt.figure(figsize=(20,25))
plt.show()
```

```python
sns.boxplot( data['Weight'] ) #show box plot of range of all vehicles weights
plt.title('Range of all Vehicles Curb Weight')
plt.show()
```

```python
sns.barplot(x=data['MPG'], y=data['Weight'])#bar plot to show MPG and Weight comparison
plt.title('MPG and Vehicle Weight Correlation')
plt.xlabel('Vehicle MPG') #x-lable
plt.ylabel('Vehicle Weight') #y-label
plt.show()
```

```python
sns.barplot(x=data['Origin'], y=data['MPG']) #barplot for comparing car origin and MPGs
plt.title('Vehicle Origin and MPG Comparison')
plt.xlabel('Country of Origin') #x-label
plt.ylabel('MPG') #y-label
plt.show()
```

```python
sns.barplot(x=data['Origin'], y=data['Displacement']) #barplot to show comparison of engine size and country of origin
plt.title('Vehicle Origins and Engine Displacement Comparison') #title
plt.xlabel('Origin of Car') #x-axis
plt.ylabel('Engine Displacement') #y-axis
plt.show()
```

```python
sns.barplot(x=data['Origin'], y=data['Acceleration']) #parplot for showing car origin and car speed/acceleration
plt.title('Country of Origin and Acceleration Comparison') #title
plt.xlabel('Origin of Car') #x-lable
plt.ylabel('Car Acceleration') #y-label
plt.show()
```

```python
sns.barplot(x=data['Origin'], y=data['Weight'])  #Barplot to compare country of orgin and car weight
plt.title('Comparison of Vehicle Weight and Country of Origin') #title 
plt.xlabel('Origin of Car') #X-lable
plt.ylabel('Vehicle Curb Weight') #y-lable
plt.legend(loc=1)
plt.show()
```

```python
sns.displot(x=data['MPG'], y=data['Displacement']) # correlation between vehicle weight and MPGs
plt.title('Engine Displacement and MPG Correlation') #title
plt.xlabel('MPG') #Xlabel
plt.ylabel('Engine Displacement') #y-lable
plt.show()
```

```python
import matplotlib.pyplot as plt #import statement
```

```python
#create line chart
x_coords = [0, 1, 2, 3, 4] # set the values for x coordinate as a list
y_coords = [0, 3, 1, 5, 2] # set the values for y coordinate as a list
plt.plot(x_coords, y_coords) # code to create a line chart
plt.show()
```

```python
#customize the labels of tick marks of a line chart
plt.xticks([0, 1, 2, 3, 4],
           ['2016', '2017', '2018', '2019', '2020'])#set the x axis labels with xticks function
plt.yticks([0, 1, 2, 3, 4, 5],
           ['$0m', '$1m', '$2m', '$3m', '$4m', '$5m'])#set the y axis labels with yticks function
plt.plot(x_coords, y_coords) # create the line chart
plt.show()
```

```python
#Customize the chart appearance, add title etc.(we are creating the same chart again as above)
x = [0, 1, 2, 3, 4]
y = [0, 3, 1, 5, 2]
 
plt.plot(x, y, marker = 'v', linestyle = '-', linewidth = '2.5',color = 'y')#chnage the marker to 'v', change solid line to dashed line, increase the width, change the color
plt.title('Sales by Year')#add title to the chart
plt.xlabel('Year')# add label name to the x axis
plt.ylabel('Sales')#add label name to the y axis
plt.xticks([0, 1, 2, 3, 4],['2016', '2017', '2018', '2019', '2020'])#set the x axis labels with xticks function
plt.yticks([0, 1, 2, 3, 5],['$0m', '$1m', '$2m', '$3m', '$5m'])#set the y axis labels with yticks function
plt.grid(True)#code if you want to show the gridlines
#refer slides 11 and 12 for diffrent line and marker style and color 
```

```python
#create bar chart
left_edges = [0,10,20,30,40]#left edges
height = [100,200,300,400,500]#height
plt.bar(left_edges,height, width=2 , color = 'y') #code to create bar chart
plt.title('SalesbyYear')#add title to the chart
plt.xlabel('Year')# add label name to the x axis
plt.ylabel('Sales')#add label name to the y axis
plt.xticks(left_edges,['2016', '2017', '2018', '2019','2020'])#set the x axis labels with xticks function
plt.yticks(height,['$1m','$2m', '$3m', '$4m','$5m',])#set the y axis labels with yticks function
plt.grid(True)
#Refer to slide 12 for color code
```

```python
#Create pie chart
values = [20, 60, 80, 40]
plt.pie(values)
plt.show()
```

```python
#Customize the pie chart
plt.pie([20, 60, 80, 40], labels=['Q1','Q2','Q3','Q4'], explode=(0,0,0.2,0),colors=('r','y','m','b'),autopct='%1.1f%%')
plt.show()
```

```python
#Practice Problem #1
#Create the dataframe 
from matplotlib import pyplot as plt
import pandas as pd
data=pd.DataFrame({'Year':['2000','2001','2002','2003','2004','2005'],
                  'AccountingMinSalary':[85000,75000,50000,70000,80000,70500],
                  'AccountingMaxSalary':[150000,130000,80000,120000,130000,123000],
                  'EngineeringMinSalary':[75000,705000,47000,65000,81000,70000],
                  'EngineeringMaxSalary':[140000,120000,70000,110000,130100,122000]
                  }, index=['a','b','c','d','e','f'])
data
```

```python
plt.figure(figsize=(20,25))#set the size of overall figure

plt.subplot(3,2,1)#create subplot, this determines the position of the chart in the figure
plt.bar(data['Year'],data['AccountingMaxSalary'], label='Accounting Maximum Salary')
plt.bar(data['Year'],data['EngineeringMaxSalary'], label='Engineering Maximum Salary')
plt.title('Accounting vs Engineering Maximum Salary')
plt.xlabel('Year')
plt.ylabel('Maximum Salary')
plt.legend(loc='best')

plt.subplot(3,2,2)
plt.bar(data['Year'],data['AccountingMinSalary'], label='Accounting Minimum Salary')
plt.bar(data['Year'],data['EngineeringMinSalary'], label='Engineering Minimum Salary')
plt.title('Accounting vs Engineering Minimum Salary')
plt.xlabel('Year')
plt.ylabel('Minimum Salary')
plt.legend(loc='best')


plt.subplot(3,2,3)
plt.bar(data['Year'],data['AccountingMaxSalary'], label='Accounting Maximum Salary')
plt.bar(data['Year'],data['AccountingMinSalary'], label='Accounting Minimum Salary')
plt.title('Accounting max vs min Salary')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.legend(loc='best')

plt.subplot(3,2,4)
plt.bar(data['Year'],data['EngineeringMaxSalary'], label='Engineering Maximum Salary')
plt.bar(data['Year'],data['EngineeringMinSalary'], label='Engineering Minimum Salary')
plt.title('Engineering max vs min Salary')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.legend(loc='best')

plt.subplot(3,2,5)
plt.pie(data['AccountingMaxSalary'], labels=data['Year'],colors=('r','y','m','b'))
plt.title('Pie chart for Accounting maximum salary')
plt.legend(loc='best')
```
